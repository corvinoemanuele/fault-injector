```python
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
import pandas as pd

ea = EventAccumulator('runs/CIFAR100/RESNET56/May07_23-07-00/')
ea.Reload()

# estrai tutte le metriche
df_test_acc = pd.DataFrame(ea.Scalars('Accuracy/test'))
df_train_acc = pd.DataFrame(ea.Scalars('Accuracy/train'))
df_test_loss = pd.DataFrame(ea.Scalars('Loss/test'))
df_train_loss = pd.DataFrame(ea.Scalars('Loss/train'))

# unisci in un unico DataFrame
df = pd.DataFrame({
    'epoch': df_test_acc['step'],
    'test_acc': df_test_acc['value'],
    'train_acc': df_train_acc['value'],
    'test_loss': df_test_loss['value'],
    'train_loss': df_train_loss['value']
})

# salva come CSV
df.to_csv('metrics_resnet56_lr0001.csv', index=False)
```
